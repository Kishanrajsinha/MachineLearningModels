# STEP 1 : preprocessing data

```python
df.dropna(axis=0, how='all')
```

# Step 2 : Feature vs Labels

```python
features = df
```

```python
features = features.drop('Date', axis=1)
features = features.drop('Time', axis=1)
features = features.drop('C6H6(GT)', axis=1)
features = features.drop('PT08.S4(NO2)', axis=1)
```

```python
labels = df['C6H6(GT)'].values
```

```python
features = features.values
```

# Step 3 : Train and  test portion

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.3)
```

```python

print("X_trian shape --> {}".format(X_train.shape))
print("y_train shape --> {}".format(y_train.shape))
print("X_test shape --> {}".format(X_test.shape))
print("y_test shape --> {}".format(y_test.shape))
```

# Step 4. Regression

# # LINEAR

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
plt.plot(y_test,regressor.predict(X_test))
plt.show()
```

```python
plt.scatter(regressor.predict(X_test),y_test)
plt.show()
```
